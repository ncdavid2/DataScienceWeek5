```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df2 = pd.read_csv('../DataScienceWeek5/Data/ch1_discoveries.csv')

df2['date'] = pd.to_datetime(df2['date'])
print(df2.dtypes)
```

```python
df2 = df2.set_index('date')
df2.plot()
plt.show()
```

```python
plt.style.use('ggplot')
df2_sublet = df2['01-01-1860':'01-01-1885']
ax = df2_sublet.plot(color='blue',fontsize=14, linestyle='--')
ax.axhline(y=4, color='g', linestyle='-')
ax.axhline(y=6, color='r', linestyle='-', linewidth=6)
ax.set_title('Time series', fontsize=16)
ax.set_ylabel('Number of discoveries', fontsize=16)
ax.set_xlabel('Date', fontsize=16)
plt.show()

plt.show()
```

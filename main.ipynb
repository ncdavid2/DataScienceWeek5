```python
df = pd.read_csv('../DataScienceWeek5/Data/ch2_co2_levels.csv')
df['datestamp'] = pd.to_datetime(df['datestamp'], dayfirst=False)
df = df.set_index('datestamp')
plt.style.use('dark_background')
df.plot()
plt.show()
```

```python
ax = df.plot(color = 'blue')
ax.set_xlabel('Date')
ax.set_ylabel('The value of my Y axis')
ax.set_title('The title of my plot')
plt.show()
```

```python
ax = df.plot(figsize = (20,5), fontsize=12,
linewidth=3, linestyle='--')
ax.set_xlabel('Date', fontsize=16)
ax.set_ylabel('The value of my Y axis', fontsize=16)
ax.set_title('The title of my plot', fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
df_subset = df['1960' : '1970']
ax = df_subset.plot(color='blue', fontsize=14)
plt.show()
```
